```python
def stone_changes(input_number):
    number = int(input_number)
    if number == 0:
        output = '1'

    elif len(str(input_number)) % 2 == 0:
        output = str(int(input_number[:len(input_number)//2])),' ', str(int(input_number[len(input_number)//2:]))
        
    else:
        output = str(number*2024)
        
    return list(output)
```

```python
test = '125 17'
new_seq=[]
for elt in test.split():
    new_seq = new_seq + stone_changes(elt) + [' ']

result = ''.join(new_seq[:-1])
```

```python
def blink(stones_dict):
    new_dict = defaultdict(lambda: 0)
    for key, value in stones_dict.items():
    
        if key == '0':
            new_dict['1'] += value

        elif len(str(key)) % 2 == 0:

            new_dict[str(int(str(key)[:len(str(key))//2]))] += value 
            new_dict[str(int(str(key)[len(str(key))//2:]))] += value 
            
        else:
            new_dict[str(int(key)*2024)]+=value
        
    return new_dict
```

```python
from collections import defaultdict
with open("day_11.txt") as f:
    stones = [int(i) for i in f.read().split(" ")]
stones_dict = {stone:stones.count(stone) for stone in stones}
```

```python
for _ in range(25):
    stones_dict = blink(stones_dict)
    print(sum(stones_dict.values()))
```
